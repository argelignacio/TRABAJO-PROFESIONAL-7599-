```python
import os
import re

EXPERIMENT_ID = 'e076a030'
BASE_FOLDER = '../results/'
experiment_folders = sorted([folder for folder in os.listdir(BASE_FOLDER) if re.match('.*' + EXPERIMENT_ID, folder)])
experiment_folders
```

```python
import pandas as pd

df = pd.read_csv(BASE_FOLDER + experiment_folders[0] + '/df_result.csv')
df
```

```python
df = pd.concat([
    pd.read_csv(BASE_FOLDER + folder + '/df_result.csv') for folder in experiment_folders
])
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df.sort_values(by='Precision_reduced_hdbscan_custom_64', ascending=False)
```

```python
df.fillna(0, inplace=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Comparo los reducidos solamente
sample = df[[column for column in df.columns if 'reduced_hdbscan_custom' in column]].sort_values(by='Precision_reduced_hdbscan_custom_64', ascending=False)

# Los nombres salen del config.ini, refieren a:
# TamaÃ±o de Cluster - Intern Proba - Floating Nodes Proportion
# np.array para poder ordenar por una columna y que el grÃ¡fico quede mÃ¡s legible
labels = np.array([
    'Small Strong Clean',
    'Medium Mid Clean',
    'Medium Weak Noisy',
    'Medium Strong Clean',
    'Medium Mid Noisy',
    'Medium Weak Noisiest',
    'Big Strong Clean',
    'Big Mid Noisy',
    'Big Weak Noisiest'
])

plt.figure(figsize=(12, 8))
plt.title('ComparaciÃ³n de precisiÃ³n reducida para distintas task segÃºn el tamaÃ±o del embedding (HDBSCAN)')
for column in sample.columns:
    plt.plot(sample[column].values)
plt.legend(['Embedding de {}'.format(2**(len(sample.columns) - i)) for i in range(len(sample.columns))])
plt.xticks(ticks=range(len(sample)), labels=labels[sample.index], rotation=45, horizontalalignment='right')
plt.show()
```
